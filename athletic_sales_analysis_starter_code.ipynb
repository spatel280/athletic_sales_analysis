```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic_sales_2020_df = pd.read_csv('Resources/athletic_sales_2020.csv')
athletic_sales_2021_df = pd.read_csv('Resources/athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
athletic_sales_2020_df.head(20)
```

```python
# Display the 2021 sales DataFrame
athletic_sales_2021_df
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
athletic_sales_2020_df.info()
```

```python
# Check the 2021 sales data types.
athletic_sales_2021_df.info()
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
athletic_sales_combined_df = pd.concat([athletic_sales_2020_df, athletic_sales_2021_df], axis = "rows", join = "inner").reset_index(drop=True)
athletic_sales_combined_df
```

```python
# Check if any values are null.
athletic_sales_combined_df.isnull().any()
```

```python
# Check the data type of each column
athletic_sales_combined_df.info()
```

```python
# Convert the "invoice_date" to a datetime datatype
athletic_sales_combined_df['invoice_date'] = athletic_sales_combined_df.loc[:, 'invoice_date'].astype('datetime64[ns]')
```

```python
# Confirm that the "invoice_date" data type has been changed.
athletic_sales_combined_df.info()
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number of products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
athletic_sales_groupby_products_rgstct = athletic_sales_combined_df.groupby(['region','state','city'])[['units_sold']].sum()\
                                .rename(columns={'units_sold':'Total_Products_Sold'})


# Show the top 5 results.
athletic_sales_groupby_products_rgstct.sort_values(by='Total_Products_Sold', ascending=False).head(5)
```

#### Using `pivot_table`

```python
# Show the number of products sold for region, state, and city.
athletic_sales_pivottable_products_rgstct = athletic_sales_combined_df.pivot_table(values = 'units_sold',\
                                    index = ['region', 'state', 'city'], aggfunc = 'sum')

# Rename the "units_sold" column to "Total_Products_Sold"
athletic_sales_pivottable_products_rgstct = athletic_sales_pivottable_products_rgstct.rename(columns={'units_sold' : 'Total_Products_Sold'})

# Show the top 5 results.
athletic_sales_pivottable_products_rgstct.sort_values(by='Total_Products_Sold', ascending=False).head(5)
```

### 3. Determine which Region had the Most Sales

#### Using `groupby`

```python
# Show the total sales for the products sold for each region, state, and city.
# Rename the "total_sales" column to "Total_Sales"
athletic_sales_groupby_sales_rgstct = athletic_sales_combined_df.groupby(['region','state','city'])[['total_sales']].sum()\
                                        .rename(columns={'total_sales':'Total_Sales'})

# Show the top 5 results.
athletic_sales_groupby_sales_rgstct.sort_values(by='Total_Sales', ascending=False).head(5)
```

#### Using `pivot_table`

```python
# Show the total sales for the products sold for each region, state, and city.
athletic_sales_pivottable_sales_rgstct = athletic_sales_combined_df.pivot_table(values = 'total_sales',\
                                    index = ['region', 'state', 'city'], aggfunc = 'sum')

# Optional: Rename the "total_sales" column to "Total Sales"
athletic_sales_pivottable_sales_rgstct = athletic_sales_pivottable_sales_rgstct.rename(columns={'total_sales' : 'Total_Sales'})

# Show the top 5 results.
athletic_sales_pivottable_sales_rgstct.sort_values(by='Total_Sales', ascending=False).head(5)
```
